```python
#Import the libraries
import pandas as pd
import numpy as np
```

```python
# Import the Erbil Univ Medical dataset dataset
df = pd.read_csv("ErbilUnivMedicaldataset.csv")
# Print dataframe first few rows
df.head()
```

```python
# Print dataframe info
df.info()
```

```python
# Convert glucose column to integer
df['glucose'] = df['glucose'].astype(np.int64)
```

```python
# Convert heartattack labels to 0,1
df['heartattack'].replace({'negative': 0, 'positive': 1, 'positive            ': 1}, inplace=True)
# Convert heartattack column to integer
df['heartattack'] = pd.to_numeric(df['heartattack'])
```

```python
# Print dataframe first few rows
df.head()
```

```python
# Print dataframe info
df.info()
```

```python
from sklearn.model_selection import train_test_split
# Do Train Test split 80:20
X = df.loc[:, df.columns != 'heartattack']
y = df['heartattack']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

Classification Algo 1: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Logistic regression classifier
clf = LogisticRegression(C=10.0, max_iter=500).fit(X_train, y_train)
```

```python
# Do predictions on X_test using Logistic regression
predictions = clf.predict(X_test)
predictions
```

```python
# Add actual Target value to X_test
X_test['heartattack'] = y_test
X_test.head()
```

```python
# Add predicted Target value to X_test
X_test['Prediction'] = predictions
X_test.head()
```

```python
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix
labels = np.unique(y_test)
cm = confusion_matrix(y_test, predictions, labels=labels)
#cm

pd.DataFrame(cm, index=labels, columns=labels)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
tn, fp, fn, tp
```

```python
from sklearn.metrics import classification_report

# Print classification report
print(classification_report(y_test, predictions))
```

```python
from sklearn import metrics

# Calculate Classification metrics: Accuracy, Precision, Recall
accuracy = metrics.accuracy_score(y_test, predictions)
Precision = metrics.precision_score(y_test, predictions)
Recall = metrics.recall_score(y_test, predictions)
print("Accuracy :",accuracy)
print("Precision:",Precision)
print("Recall   :",Recall)
```

```python
from sklearn.model_selection import GridSearchCV

# Use GridSearchCV with Logistic Regression to recommend best parameters

# Define parameter grid
param_grid = {'C': [0.01, 0.1, 1, 10], 'solver': ['lbfgs'], 'max_iter' : [500, 1000, 5000, 10000]}

# Initialize the grid search model
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5)

# Fit the grid search model
grid_search.fit(X_train, y_train)

# Print best parameters
print("Best parameters:", grid_search.best_params_)
```
